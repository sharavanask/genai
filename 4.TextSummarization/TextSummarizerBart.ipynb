```python
s=""
doc=fitz.open("data.pdf")
for i in doc:
  s+=i.get_text()

s
```

```python
!pip install transformers
from transformers import BartForConditionalGeneration, BartTokenizer
```

```python
model="facebook/bart-large-cnn"
tokenizer=BartTokenizer.from_pretrained(model)
model=BartForConditionalGeneration.from_pretrained(model)
```

```python
import textwrap
inputs = tokenizer.encode(s, return_tensors="pt", max_length=1024, truncation=True)
summary_ids = model.generate(inputs, max_length=250, min_length=100, length_penalty=1.0, num_beams=4, early_stopping=True)

summary = tokenizer.decode(summary_ids[0], skip_special_tokens=True)
formatted_summary = "\n".join(textwrap.wrap(summary, width=80))
formatted_summary
```
